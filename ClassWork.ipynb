Part 1: The Automated Summary (Coding)



```python
import numpy as np

# Initial patient recovery scores
data = [4.2, 5.1, 3.8, 7.2, 5.5, 6.0]

def summarize_clinic(patient_data):
    mean_val = np.mean(patient_data)
    std_dev = np.std(patient_data)
    return mean_val, std_dev

# 1. Run the initial summary
mean, std = summarize_clinic(data)
print(f"Initial - Mean: {mean:.2f}, Std Dev: {std:.2f}")

# 2. Logic Check: Adding the outlier (15.0)
data.append(15.0)
new_mean, new_std = summarize_clinic(data)
print(f"With Outlier - Mean: {new_mean:.2f}, Std Dev: {new_std:.2f}")
```

Part 2: Visualizing the "Compass"

```python
import matplotlib.pyplot as plt

# Lab Data: Dosage vs Recovery
dosage = [10, 20, 30, 40, 50]
recovery = [2.1, 4.5, 5.9, 8.2, 9.7]

plt.figure(figsize=(10, 6))

# 1. Create Scatter Plot of actual data
plt.scatter(dosage, recovery, color='black', label='Actual Patient Data', zorder=5)

# 2 & 3. Developer Challenge: Testing different weights
weights = [0.1, 0.2, 0.5, 0.8]
colors = ['red', 'orange', 'green', 'blue']

for w, color in zip(weights, colors):
    # Prediction formula: Recovery = Weight * Dosage
    prediction = [w * d for d in dosage]
    plt.plot(dosage, prediction, label=f'Weight w={w}', color=color, linestyle='--')

# Formatting
plt.title("Clinical Dashboard: Prediction vs. Reality")
plt.xlabel("Dosage (mg)")
plt.ylabel("Recovery Score")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
