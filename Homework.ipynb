Task 1 Code

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Setup: Imagine these are Calories vs. Exercise Time ---
def calculate_loss(w):
    # This represents our "U-shaped" valley. The bottom (best score) is at 0.
    return w**2

def calculate_gradient(w):
    # The "Compass" - derivative of w^2 is 2*w
    return 2 * w

def train_model(learning_rate, iterations=50):
    weight = 10.0  # Start far away from the bottom
    loss_history = []

    for i in range(iterations):
        loss = calculate_loss(weight)
        loss_history.append(loss)

        # --- THE FIX: Subtract the gradient to go DOWN the hill ---
        gradient = calculate_gradient(weight)
        weight = weight - (learning_rate * gradient)

    return loss_history

# --- Experiment: Testing Two Learning Rates ---
slow_rate = 0.01
fast_rate = 0.1

history_slow = train_model(slow_rate)
history_fast = train_model(fast_rate)

# --- Visualizing the Results ---
plt.figure(figsize=(10, 6))
plt.plot(history_slow, label=f"Slow Rate ({slow_rate})", color='red', linewidth=2)
plt.plot(history_fast, label=f"Fast Rate ({fast_rate})", color='green', linewidth=2)

plt.title("The Loss Curve: How Fast Does the AI Learn?")
plt.xlabel("Iteration (Step Number)")
plt.ylabel("Loss (Error Amount)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Setup our "Loss" (The Valley) and "Gradient" (The Compass)
def calculate_loss(w):
    return w**2  # We want w to reach 0 for the lowest error

def calculate_gradient(w):
    return 2 * w  # The slope of the line

def train_simulation(lr, iterations=100):
    weight = 10.0  # Start high up the hill
    loss_history = []
    for i in range(iterations):
        loss = calculate_loss(weight)
        loss_history.append(loss)

        # FIX: Subtract instead of Add
        gradient = calculate_gradient(weight)
        weight = weight - (lr * gradient)
    return loss_history

# 2. Run experiments with your specific rates
lr_too_slow = 0.000001
lr_converged = 0.1

results_slow = train_simulation(lr_too_slow)
results_fast = train_simulation(lr_converged)

# 3. Create the Visualization (Loss Curve)
plt.figure(figsize=(10, 5))
plt.plot(results_slow, label=f"Learning Rate {lr_too_slow} (Too Slow)", color='red')
plt.plot(results_fast, label=f"Learning Rate {lr_converged} (Converged)", color='green')

plt.title("Impact of Step Size on Model Learning")
plt.xlabel("Iteration (Time)")
plt.ylabel("Loss (Error)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
